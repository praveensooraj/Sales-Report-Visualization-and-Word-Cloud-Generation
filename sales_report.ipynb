```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a=('AMOUNT_AFTER_TAX')#which customer paid more amount
b=data.groupby('CLIENT').sum().sort_values('AMOUNT_AFTER_TAX',ascending=False)[['AMOUNT_AFTER_TAX']][0:25]
sns.barplot(x=a,y=b.index,data=b)
```

```python
c='AMOUNT_AFTER_TAX'#which product is highly taxed
d=data.groupby('PRODUCT').sum().sort_values('AMOUNT_AFTER_TAX',ascending=False)[['AMOUNT_AFTER_TAX']][0:25]
sns.barplot(x=c,y=d.index,data=d)
```

```python
e='QUANTITY'#which product is bought more
f=data.groupby('PRODUCT').sum().sort_values('QUANTITY',ascending=False)[['QUANTITY']][0:25]
sns.barplot(x=e,y=f.index,data=f)
```

```python
g='AMOUNT_AFTER_TAX'#on which date sales was more
h=data.groupby('DATE').sum().sort_values('AMOUNT_AFTER_TAX',ascending=False)[['AMOUNT_AFTER_TAX']][0:25]
sns.barplot(x=g,y=h.index,data=h)
```

```python
from wordcloud import WordCloud
```

```python
text = ' '.join(data['PRODUCT'].dropna().astype(str).values)
wordcloud=WordCloud(background_color='white',width=800,height=400).generate(text)
fig,ax=plt.subplots(figsize=(15,6))
ax.imshow(wordcloud,interpolation='bilinear')
ax.set_title('word cloud of PRODUCT column')
ax.set_xticks([])
ax.set_yticks([])
plt.show()
```
